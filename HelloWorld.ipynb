```python
from transformers import pipeline

# Example: Create a text generation pipeline
generator = pipeline('text-generation', model='gpt2')
result = generator("Once upon a time, in a land", max_length=50, num_return_sequences=1)
print(result)
print(result[0]['generated_text'])
```

Text model generation with WebUI

```python
from transformers import pipeline
import gradio as gr


def genText(txt):
  # Example: Create a text generation pipeline
  generator = pipeline('text-generation', model='gpt2')
  result = generator(txt, max_length=50, num_return_sequences=1)
  print(result)
  print(result[0]['generated_text'])
  return result[0]['generated_text']




# Create a Gradio interface
interface = gr.Interface(
    fn=genText,
    inputs="text",
    outputs='text'

)

interface.launch()
```
